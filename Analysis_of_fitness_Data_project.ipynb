##Task 2

```python
#List of columns to drop
cols_to_drop = ['Route Name',"Friend's Tagged",'Notes','Activity Id','GPX File']  # Add the names of the columns you want to drop

# Drop unnecessary columns
df_activities.drop(columns=cols_to_drop, inplace=True)

# Calculate activity type counts
activity_type_counts = df_activities['Type'].value_counts()

# Rename 'Other' values to 'Unicycling' in Type column
df_activities['Type'] = df_activities['Type'].str.replace('Other', 'Unicycling')

# Count missing values in each column
missing_values_count = df_activities.isnull().sum()

# Print the activity type counts
print("Activity Type Counts:")
print(activity_type_counts)

# Print missing values count
print("\nMissing Values Count:")
print(missing_values_count)
```

# Task 3

```python
# Calculate the sample mean for 'Cycling' activity type
avg_hr_cycle = df_activities[df_activities['Type'] == 'Cycling']['Average Heart Rate (bpm)'].mean()

# Filter df_activities for 'Cycling' activity type
df_cycle = df_activities[df_activities['Type'] == 'Cycling'].copy()

# Fill missing values for Average Heart Rate (bpm) in df_cycle
df_cycle['Average Heart Rate (bpm)'].fillna(int(avg_hr_cycle), inplace=True)

# Count missing values for all columns in df_cycle
missing_values_count_cycle = df_cycle.isnull().sum()

# Print the average heart rate for 'Cycling'
print("Average Heart Rate for Cycling:", avg_hr_cycle)

# Print missing values count for df_cycle
print("\nMissing Values Count for df_cycle:")
print(missing_values_count_cycle)
```

# Task 4

```python

# Subset df_run for data from 2015 through 2018
runs_subset_2015_2018 = df_activities['2015':'2018']

# Calculate annual averages using resample() with 'A' alias
annual_means = runs_subset_2015_2018.resample('A').mean()

# Calculate average weekly statistics using resample() with 'W' alias and mean() method twice
weekly_means = runs_subset_2015_2018.resample('W').mean()
weekly_means_average = weekly_means.resample('W').mean()

# Filter the 'Distance (km)' column and count the average number of trainings per week
weekly_counts_average = df_activities['Distance (km)'].resample('W').agg(['count', 'mean'])

# Print the results
print("Annual Averages:")
print(annual_means)

print("\nAverage Weekly Statistics:")
print(weekly_means_average)

print("\nAverage Weekly Counts of Trainings:")
print(weekly_counts_average)
```

# Task 6

```python
# Subset df_run for data from 2015 through 2018
runs_subset_2015_2018 = df_activities['2015':'2018']

# Select information for distance and heart rate
runs_distance = runs_subset_2015_2018['Distance (km)']
runs_hr = runs_subset_2015_2018['Average Heart Rate (bpm)']

# Create subplots with shared x-axis
fig, (ax1, ax2) = plt.subplots(2, sharex=True, figsize=(12, 8))

# Plot distance on the first subplot
ax1.plot(runs_distance.index, runs_distance, color='green', label='Distance')
ax1.set_ylabel('Distance (km)')
ax1.legend()

# Add a horizontal line for average heart rate on the second subplot
avg_hr = runs_hr.mean()
ax2.axhline(avg_hr, color='blue', linewidth=1, linestyle='-.', label='Avg Heart Rate')
ax2.plot(runs_hr.index, runs_hr, color='red', label='Heart Rate')
ax2.set_ylabel('Heart Rate (bpm)')
ax2.set_xlabel('Date')
ax2.legend()

plt.tight_layout()
plt.show()
```

## Task 7

```python
# Subset df_run for data from 2013 through 2018
df_run_subset_2013_2018 = df_activities['2013':'2018']

# Count annual totals for Distance (km)
df_run_dist_annual = df_run_subset_2013_2018['Distance (km)'].resample('A').sum()

# Create the plot
plt.figure(figsize=(8.0, 5.0))
plt.plot(df_run_dist_annual.index, df_run_dist_annual, color='blue', marker='o', linestyle='-', linewidth=2, markersize=6)
plt.axhspan(0, 800, color='red', alpha=0.2)
plt.xlabel('Year')
plt.ylabel('Total Distance (km)')
plt.title('Annual Total Distance Run (2013-2018)')
plt.grid(True)
plt.show()
```

# Task 8

```python
import statsmodels.api as sm
# Subset df_run for data from 2013 through 2018
df_run_subset_2013_2018 = df_activities['2013':'2018']

# Resample weekly and fill NaN values using bfill()
df_run_dist_wkly = df_run_subset_2013_2018['Distance (km)'].resample('W').bfill()

# Create a plot for observed distance and decomposed trend
plt.figure(figsize=(12, 5))

# Plot observed distance
plt.plot(df_run_dist_wkly.index, df_run_dist_wkly, label='Observed Distance (Weekly)', color='blue')

# Decompose and plot trend using statsmodels' seasonal decomposition
decomposition = sm.tsa.seasonal_decompose(df_run_dist_wkly, model='additive')
plt.plot(decomposition.trend.index, decomposition.trend, label='Trend', color='orange')

plt.xlabel('Date')
plt.ylabel('Distance (km)')
plt.title('Observed Distance of Runs and Decomposed Trend')
plt.legend()
plt.grid(True)
plt.show()
```

## Task 9

```python
# Subset df_run from March 2015 through 2018
df_run_subset_march_2015_2018 = df_activities['2015-03':'2018']

# Select Average Heart Rate (bpm) column
df_run_hr_all = df_run_subset_march_2015_2018['Average Heart Rate (bpm)']

# Create a plot with customized x-axis ticks
fig, ax = plt.subplots(figsize=(8, 5))
zone_names = df_run_hr_all.index.strftime('%b %Y')
ax.plot(df_run_hr_all.index, df_run_hr_all, color='purple', marker='o', linestyle='-', linewidth=2, markersize=6)
ax.set_xticklabels(zone_names, rotation=-30, ha='left')

ax.set_xlabel('Date')
ax.set_ylabel('Average Heart Rate (bpm)')
ax.set_title('Average Heart Rate (bpm) Over Time')
ax.grid(True)
plt.tight_layout()
plt.show()
```

# Task 10

```python
#Calculate the sample mean for 'walk' activity type
avg_hr_walk = df_activities[df_activities['Type'] == 'Walking']['Average Heart Rate (bpm)'].mean()

# Filter df_activities for 'Cycling' activity type
df_walk = df_activities[df_activities['Type'] == 'Walking'].copy()

# Fill missing values for Average Heart Rate (bpm) in df_cycle
if pd.notnull(avg_hr_walk):
    df_walk['Average Heart Rate (bpm)'].fillna(int(avg_hr_walk), inplace=True)

# Count missing values for all columns in df_activities
missing_values_count_activities = df_activities.isnull().sum()

# Print the average heart rate for 'walk'
print("Average Heart Rate for Walking:", avg_hr_walk)

# Print missing values count for df_activities
print("\nMissing Values Count for df_activities:")
print(missing_values_count_activities)
```

```python
# Concatenate df_run, df_walk, and df_cycle using append()
df_run_walk_cycle = df_activities.append([df_walk, df_cycle])

# Sort the concatenated DataFrame based on index in descending order
df_run_walk_cycle.sort_index(ascending=False, inplace=True)

# Define the columns to include in the summary report
dist_climb_cols = ['Distance (km)', 'Climb (m)']

# Group by activity type and calculate sums for selected columns
df_totals = df_run_walk_cycle.groupby('Type')[dist_climb_cols].sum()

# Reshape the summary report using stack()
df_summary = df_totals.stack()

# Print the summary report
print(df_summary)
```
